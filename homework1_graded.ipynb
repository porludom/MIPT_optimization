As a result, to obtain matrice with needed L and $\mu$ we need only the list of eigenvalues. This list has length d (dimension)

__Step 1__. Generate a d number of eigenvalues in range from mu to L\
__Step 2__. Generate random orthonormal matrice of dimension d by d\
__Step 3__. Using eigenvalue decomposition, create matrice $A = QSQ^T$, where Q is generated orthonormal matrice, S is diagonal matrice, where diagonal enties are eigenvalues we have specified.
```python
import numpy as np
import scipy.linalg as la
from scipy.stats import ortho_group
import matplotlib.pyplot as plt
!pip install seaborn
#import seaborn as sns
from IPython.display import clear_output
%matplotlib inline
#sns.set_theme(style="darkgrid")
import warnings
warnings.filterwarnings("ignore")
clear_output()
```

```python
def generate_A(d : int, mu : int, L: int):
    eigens = [mu] # a list of eigenvalues. the smallest is the mu the biggest is L
    for i in range(d-2):
        eigens.append(np.random.randint(mu, L+1))
    eigens.append(L)
    eigens = sorted(eigens)
    s = np.diag(eigens)
    q = ortho_group.rvs(d) # orthonormal matrice of dimension d
    A = q @ s @ q.T # simple eigenvalue decomposition. Generates symmetric matrice
    # Now we have matrice A with specified eigenvalues
    return A
A = generate_A(5,2,6)
eigenvalues, _ = np.linalg.eig(A)
print("Eigenvalues:", eigenvalues)
print("Generated matrice:\n", A)
```

__d). (5 points)__  Implement the gradient descent method:
\begin{equation}
x_{k+1} = x_k - \gamma_k \nabla f(x_k),
\end{equation}
for a quadratic function with the matrix generated above and an arbitrary vector $b$ (preferably non-zero). It is already desirable to implement both the method and the gradient calculation as separate functions, with the method taking the target function itself (or its gradient) as a parameter. In the future it will be easier to change the target function in the minimization problem and test different methods.

Optionally: you can use the following variant of the function description.

```python
def gradient_descent(f, nabla_f, x_0, x_sol, gamma_k, type_of_step_f : str, K = 10**3, eps = 10**-5, mode = 'x_k - x^*', **args):
    '''
        f - target function
        nabla_f - gradient of the target function
        x_0 - start point
        x_sol - exact solution (it is needed for error calculation)
        gamma_k - function for calculating the method step
        type_of_step_f - way to compute step size. Variants:
                'L' : uses L and some number in numerator to compute constant step size. L and num are taken from args
                'SDIN' : uses current iteration and two constants:alpha and delta from args to compute step size.
                'BCD': The best choice descent
                'PS': Polyak-Shore step. Alpha is taken from args
                
                Note than function that use specific parameters is still determined by gamma_k parameter, they are not initialized inside here
                
        K - number of iterations (by default 1e3)
        eps - accuracy (by default 1e-5)
        mode - convergence criterion 
               Values are either 'x_k - x^*' - then the convergence criterion will be ||x_k - x^*||^2,
               or 'f(x_k) - f(x^*)' - then the convergence criterion will be f(x_k) - f(x^*),
               or 'x_k+1 - x_k', or 'f(x_k+1) - f(x_k)' (the criteria will be similar)

        The function returns the point at which the minimum is reached and the error vector
    '''
    f_opt = f(x_sol)
    x_k = x_0 # column vector
    x_kp1 = None
    errors = []
    for k in range(1, K+1):
        grad = nabla_f(x_k)
        match type_of_step_f:
            case "L":
                step = gamma_k(args["num"], args["L"])
            case "SDIN":
                step = gamma_k(args["alpha"], args["delta"], k)
            case "BCD":
                step = gamma_k(grad)
            case "PS":
                f_cur = f(x_k) # at previous
                step = gamma_k(f_cur, f_opt, args["alpha"], grad)
            case _:
                raise Exception ("no such step function type")
        x_kp1 = x_k - step * grad
        match mode:
            case "x_k - x^*":
                error = np.linalg.norm(x_kp1 - x_sol)**2
            case "f(x_k) - f(x^*)":
                error = f(x_kp1) - f(x_sol)
            case "x_k+1 - x_k":
                error = np.linalg.norm(x_kp1 - x_k)**2
            case "f(x_k+1) - f(x_k)":
                error = f(x_kp1) - f(x_k)
            case _:
                raise Exception ("no such convergence criterion")
        errors.append(error)
        if error <= eps:
            return x_kp1, errors
        x_k = x_kp1
        x_kp1 = None
    return x_k, errors
```
